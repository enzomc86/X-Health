```python
# Aqui optei pela utilzação de sklearn porque tenho mais familiaridade, além do dataset não ser muito complexo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import joblib

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
# Carregar o dataset e exibir principais métricas
df = pd.read_csv('./exercicio_DSteam_shared4-master/_data/dataset_2021-5-26-10-14.csv', 
                 sep='\t', encoding='utf-8')

print(f"Estruturado dataset: {df.shape}")
print(f"Taxa de default: {df['default'].mean():.2%}")

df = df.replace("missing", np.nan)

print("\nDados faltantes por coluna:")
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])
```

```python
# Criação de variáveis com base na analise exploratória

#para indicar se tem restrições (protestoa/ações judiciais/dívidas)
df['tem_restricoes'] = ((df['quant_protestos'] > 0) | 
                        (df['quant_acao_judicial'] > 0) | 
                        (df['dividas_vencidas_qtd'] > 0)).astype(int)

#Razão entre valor vencido e valor total do pedido
df['razao_vencido_total'] = df['valor_vencido'] / (df['valor_total_pedido'] + 1)

#Para dados ausentes em forma de pagamento 
df['forma_pagamento_missing'] = df['forma_pagamento'].isnull().astype(int)



print(f"\nnova variável 'tem_restricoes':")
print(df['tem_restricoes'].value_counts())
```

```python
# Selecionar as variáveis que julguei mais importantes na EDA, juntando e verificando se temos valores nullos nelas
features_numericas = [
    'ioi_36months',          
    'ioi_3months',            
    'valor_total_pedido',  
    'valor_quitado',        
    'valor_vencido',         
    'quant_protestos',       
    'razao_vencido_total'    
]

features_binarias = [
    'tem_restricoes',          
    'forma_pagamento_missing'   
]


features_finais = features_numericas + features_binarias

```

```python
# Preparar X (features) e y (target)
X = df[features_finais]
y = df['default']

# Remover linhas com valores nulos 
mask_validos = ~X.isnull().any(axis=1)
X = X[mask_validos]
y = y[mask_validos]

# Dividir em treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Dados de treino: {X_train.shape[0]:,} registros")
print(f"Dados de teste: {X_test.shape[0]:,} registros")
print(f"Taxa de default no treino: {y_train.mean():.2%}")
print(f"Taxa de default no teste: {y_test.mean():.2%}")
```

```python
#Treinei dois modelos regressão logistica x random forest para comparar o melhor resultado

#normalização
pipeline_logistico = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic', LogisticRegression(random_state=42, max_iter=1000))
])
pipeline_logistico.fit(X_train, y_train)

modelo_forest = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_forest.fit(X_train, y_train)


modelo_logistico = pipeline_logistico

print("1. Regressão Logística")
print("2. Random Forest")
```

```python
# Função para avaliar modelo
def avaliar_modelo(modelo, nome):
    # Fazer predições
    pred_treino = modelo.predict(X_train)
    pred_teste = modelo.predict(X_test)
    # Calcular AUC (métrica principal de funcionalidade)
    prob_teste = modelo.predict_proba(X_test)[:, 1]  # Probabilidade da classe 1 (default)
    auc = roc_auc_score(y_test, prob_teste)
    print(f"AUC-ROC: {auc:.3f}")
    print(classification_report(y_test, pred_teste))
    return auc, prob_teste

auc_logistico, prob_logistico = avaliar_modelo(modelo_logistico, "Regressão Logística")
auc_forest, prob_forest = avaliar_modelo(modelo_forest, "Random Forest")

# Escolher o melhor modelo
if auc_logistico > auc_forest:
    melhor_modelo = modelo_logistico
    melhor_nome = "Regressão Logística"
    melhores_probs = prob_logistico
else:
    melhor_modelo = modelo_forest
    melhor_nome = "Random Forest"
    melhores_probs = prob_forest

print(melhor_nome)
```

```python
# Importância das variáveis dentro do modelo random forest
importancias = modelo_forest.feature_importances_

df_importancia = pd.DataFrame({
    'Variavel': X_train.columns,
    'Importancia': importancias
}).sort_values('Importancia', ascending=False)

print(df_importancia.round(3))

plt.figure(figsize=(10, 6))
plt.barh(df_importancia['Variavel'], df_importancia['Importancia'])
plt.title('Importância das Variáveis no Random Forest')
plt.xlabel('Importância')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Probabilidades previstas
probs = modelo_forest.predict_proba(X_test)[:,1]

# Distribuição das probabilidades
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

axes[0].hist(probs[y_test == 0], bins=30, alpha=0.7, label='Classe 0', color='green')
axes[0].hist(probs[y_test == 1], bins=30, alpha=0.7, label='Classe 1', color='red')
axes[0].set_title('Distribuição das Probabilidades')
axes[0].set_xlabel('Probabilidade prevista')
axes[0].set_ylabel('Frequência')
axes[0].legend()

# Matriz de confusão
pred = modelo_forest.predict(X_test)
cm = confusion_matrix(y_test, pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[1])
axes[1].set_title('Matriz de Confusão')
axes[1].set_xlabel('Predito')
axes[1].set_ylabel('Real')

plt.tight_layout()
plt.show()

# Estatísticas
tn, fp, fn, tp = cm.ravel()
print(f"Verdadeiros Negativos (TN): {tn:,}")
print(f"Falsos Positivos (FP): {fp:,}")
print(f"Falsos Negativos (FN): {fn:,}")
print(f"Verdadeiros Positivos (TP): {tp:,}")
```

```python
# Ordenar clientes pelo risco (probabilidades mais altas primeiro)
indices_ordenados = np.argsort(probs)[::-1]
probs_ordenadas = probs[indices_ordenados]
y_ordenado = y_test.iloc[indices_ordenados].values

print(f"Total de clientes no teste: {len(y_test):,}")
print(f"Total de casos positivos (classe 1): {y_test.sum():,} ({y_test.mean():.1%})")

# Top 10% mais arriscados
top_10 = int(0.1 * len(y_test))
capturados = y_ordenado[:top_10].sum()

print("10% MAIS ARRISCADOS:")
print(f"Clientes analisados: {top_10:,}")
print(f"Casos positivos capturados: {capturados:,}")
print(f"Taxa de positivos nesse grupo: {capturados/top_10:.1%}")
print(f"% do total de positivos capturados: {capturados/y_test.sum():.1%}")

# Análise por faixas
print("FAIXAS DE RISCO:")
for p in [10, 20, 30]:
    n = int(p/100 * len(y_test))
    capt = y_ordenado[:n].sum()
    print(f"Top {p}%: {capt:,} casos positivos ({capt/n:.1%} do grupo, {capt/y_test.sum():.1%} do total)")
```
